```python
import pandas as pd 
import numpy as np 
import os
pwd = os.getcwd()
filepath = os.path.join(pwd ,"Data - Survey Monkey Output Edited.xlsx")
```

```python
dataset = pd.read_excel(filepath, sheet_name = "Edited_Data")
# dataset
```

```python
dataset_modified = dataset.copy()
# dataset_modified.columns
```

```python
columns_drop = ['Start Date', 'End Date', 'Email Address','First Name', 'Last Name', 'Custom Data 1']
dataset_modified = dataset_modified.drop(columns=columns_drop)
```

```python
len(dataset_modified)
dataset_modified
```

```python
id_vars = list(dataset_modified.columns)[0:8]
value_vars = list(dataset_modified.columns)[8:]
```

```python
dataset_melted = dataset_modified.melt(id_vars=id_vars,value_vars=value_vars,var_name="Question + Subquestion",value_name="Answer")
```

```python
question_import = pd.read_excel(filepath,sheet_name="Question")
question = question_import.copy()
```

```python
question.drop(columns=["Raw Question","Raw Subquestion","Subquestion"],inplace=True)
question
```

```python
data_merge = pd.merge(left=dataset_melted,right=question,how="left",left_on="Question + Subquestion",right_on="Question + Subquestion")
```

```python
data_merge
```

```python
Respondent = data_merge[data_merge["Answer"].notna()]
Respondent =  Respondent.groupby("Question")["Respondent ID"].nunique().reset_index()
Respondent.rename(columns={"Respondent ID":"Respondents"},inplace=True)
# Respondent
```

```python
data_merge_2 = pd.merge(left=data_merge,right=Respondent,how="left",left_on="Question",right_on="Question")
data_merge_2
```

```python
Same_Answer = data_merge#[data_merge["Answer"].notna()]
Same_Answer =  Same_Answer.groupby(["Question + Subquestion","Answer"])["Respondent ID"].nunique().reset_index()
Same_Answer.rename(columns={"Respondent ID":"Same Answer"},inplace=True)
```

```python
Same_Answer
```

```python
data_merge_3 = pd.merge(left=data_merge_2,right=Same_Answer,how="left",left_on=["Question + Subquestion","Answer"],right_on=["Question + Subquestion","Answer"])
```

```python
data_merge_3["Same Answer"].fillna(value=0,inplace=True)
```

```python
data_merge_3
```

```python
output = data_merge_3.copy()
output
```

```python
output.to_excel(pwd + "//Final_output.xlsx",index=False)
```
